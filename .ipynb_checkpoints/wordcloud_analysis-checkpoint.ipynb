```python
from collections import Counter

import nltk
import numpy as np
import pandas as pd
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt

import re

%matplotlib inline
```

```python
df_male = pd.read_excel("results/male.xlsx").dropna()
df_female = pd.read_excel("results/female.xlsx").dropna()

data_male = (df_male.text.tolist())
data_female = (df_female.text.tolist())
```

```python
# suicide
df_male_suicide = " ".join(post for post in (df_male[df_male["label"] == "suicide"])["text"])
df_female_suicide = " ".join(post for post in (df_female[df_female["label"] == "suicide"])["text"])

# get raw stats for top 25
male_suicide_counter = Counter(df_male_suicide.split())
print("===male===")
# make list of words from wordcloud
male_suicide_tops = {"never":0, "know":0, "think":0, "people":0, "time":0, "going":0, "year":0, "thing":0, "ive":0, "feel":0, "much":0, "make":0, "go": 0, "im":0, "will":0, "friend":0, "really":0, "can":0, "still":0, "one":0, "even":0, "work":0, "love":0, "now":0, "life":0}

for key in male_suicide_tops:
    male_suicide_tops[key] = male_suicide_counter[key]
# save as xlsx
pd.DataFrame.from_dict(male_suicide_tops, orient="index").to_excel("results/male_suicide_top_words.xlsx")

female_suicide_counter = Counter(df_female_suicide.split())
print("===female===")
# make list of words from wordcloud
female_suicide_tops = {"anymore":0, "really":0, "family":0, "life":0, "got":0, "thought":0, "one":0, "has":0, "help":0, "ive":0, "year":0, "friend":0, "going":0, "im":0, "go":0, "feel":0, "know":0, "even":0, "mom":0, "job":0, "tired":0, "move":0, "make":0, "time":0, "everyone":0}
for key in female_suicide_tops:
    female_suicide_tops[key] = female_suicide_counter[key]
# save as xlsx
pd.DataFrame.from_dict(female_suicide_tops, orient="index").to_excel("results/female_suicide_top_words.xlsx")


suicide_male_wordcloud = WordCloud(stopwords=STOPWORDS, background_color="white", max_words=25).generate(re.sub("pare", "", re.sub("nt", "", re.sub("wa", "", df_male_suicide))))
plt.imshow(suicide_male_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Male Suicide WordCloud", fontsize=20, color="black", loc="left")
plt.show()

suicide_female_wordcloud = WordCloud(stopwords=STOPWORDS, background_color="white", max_words=25).generate(re.sub("nt", "", re.sub("alys", "", re.sub("wa", "", df_female_suicide))))
plt.imshow(suicide_female_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Female Suicide WordCloud", fontsize=20, color="black", loc="left")
plt.show()
```
